```python
import pandas as pd
import numpy as np
```

```python
import sys
import os
os.environ["TOKENIZERS_PARALLELISM"] = "false"

src_dir = os.path.join(os.getcwd(), 'src')
sys.path.append(os.path.abspath(src_dir))
```

```python
def evaluate(df):
    results = df.groupby("evaluation")["sentence"].count()
    if len(results) == 1:
        return 0
    accuracy = results.iloc[1] / (results.iloc[0] + results.iloc[1])
    return np.round(accuracy, 2)
```

```python
def build_csv(model_name):
    data_type = ["city", "train_station"]
    versions = {
        "1": "direct routing",
        "2": "multi stop",
        "3": "exclusion routing"
    }

    results = []

    for version, tested_feature in versions.items():
        for selected_type in data_type:
            test_name = f"test{version}_{selected_type}_results"
            city_nlp_df = pd.read_csv(f'./processed/test_results/{model_name}/{test_name}.csv')
            results.append(
                {
                    "model name": model_name,
                    "data": selected_type,
                    "accuracy": evaluate(city_nlp_df),
                    "tested feature": tested_feature
                }
            )

    return pd.DataFrame(results)
```

```python
city_nlp_df = build_csv("city_nlp")
cc4000_df = build_csv("4000_classes_classification")
spacy = build_csv("spacy")
final_df = pd.concat([city_nlp_df, cc4000_df, spacy])
final_df
```
